```python
x0 = 1
N = 10000

dt = 0.01
T = 100
```

```python
time_steps = list(np.arange(dt, T, dt)) 

simulations = np.zeros((N, len(time_steps)+1)) 
simulations[:, 0] = x0 
```

$ grad(((x - 4)^2 - 2)^2) = (4(-2+(-4+x)^2)*(-4+x))= (-8+4(-4+x)^2)*(-4+x) = 4(x-4)^3 - 8(x-4)$

```python
x = simulations[:, 0]  # starting point

for i in range(1, len(time_steps) + 1):
    x_current = (4*(x - 4)**3 - 8*(x - 4))*dt
    random_n = np.sqrt(2*dt) * np.random.randn(N)
    
    x_new = x - x_current + random_n
    x = x_new
    
    simulations[:, i] = x_new 
```

```python
simulations
```

```python
import seaborn as sns
```

```python
sns.histplot(simulations[:, -1], bins=60, kde=True, alpha=0.9, label=f't = 100', stat = 'density')
plt.xlabel('Size of members')
plt.ylabel('Frequency')

plt.legend()
plt.show()
```
