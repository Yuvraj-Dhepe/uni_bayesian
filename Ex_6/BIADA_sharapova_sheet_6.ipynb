# Exercise 1.1

```python
M = 30
b = np.zeros(M)
c = np.zeros(M)
```

```python
def f_x(x):
    return np.sqrt(1-(x**2))
```

```python
# for i in range(1, M+1):
#    b[i] = 1/i
#    c[i]= ((1/(2*i))+((i-1)/i))

fM = np.zeros(M)
for j in range(1, M):
    for i in range(1, j+1):
        b[i] = 1/j
        c[i] = (1/(2*j)) + ((i-1)/j)
        fM[j] += b[i]*f_x(c[i])
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1, M), (fM[1:] - math.pi/4))
plt.title('error')
plt.xlabel("M number")
plt.ylabel("error")
plt.show()
```

# Exercise 2.3

Let M = 1,2,4,··· ,256. For each M, do N = 10000 simulations to approximate the expectation using fM. For each M, calculate the mean and the variance of fM over N rounds. Visualize your result.

```python
M = 256
N = 10000

mean = 1
sigma = np.sqrt(3)

(mean_list, E_fM_list, var_fM_list, true_variance)  = (np.zeros(M), np.zeros(M),np.zeros(M),np.zeros(M))
```

```python
def f_x(x):
  return (1+2*x+x**2)
```

```python
sample = normal(mean, sigma, N)
```

```python
for i in range(1, M+1):
    fM = np.zeros(N)

    for j in range(N):
        f_value = f_x(sample[j])
        fM[j] = (1/i) * np.sum(f_value)

    E_fM_list[i-1] = np.sum(fM)/N
    var_fM_list[i-1] = np.sum((fM - E_fM_list[i-1])**2)/(N-1)
    true_variance[i-1] = 66/i
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1, M+1), E_fM_list)
plt.title('Mean')
plt.xlabel("M number")
plt.ylabel("Expectation")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1, M+1), true_variance, label = 'true_variance')
plt.plot(range(1, M+1), var_fM_list, label = 'fM_variance')
plt.title('Variance approximation')
plt.xlabel("M number")
plt.ylabel("Variance")
plt.legend()
plt.show()
```
