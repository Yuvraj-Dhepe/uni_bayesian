## Task 1

```python
P = np.array([[1/2, 2/3, 0, 0], [1/2, 1/3, 0, 0], [0, 0, 1/4, 4/5], [0, 0, 3/4, 1/5]]) # Transition matrix
X_0 = np.array([1,0,0,0]) # first initialization
```

```python
t = 100
X_t = []

for i in range(t):
   X_0 = np.dot(P,X_0)
   X_t.append(X_0)
```

```python
X_t[-1]
```

The distribution of X_t tends towards its 1st invariant measure distribution

```python
P = np.array([[1/2, 2/3, 0, 0], [1/2, 1/3, 0, 0], [0, 0, 1/4, 4/5], [0, 0, 3/4, 1/5]]) # Transition matrix
X_0 = np.array([0,0,1,0]) # second initialization
```

```python
X_t[-1]
```
